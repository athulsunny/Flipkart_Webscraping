**Flipkart Webscraping**

```python
#Importing Required Libraries
import requests
from flask import Flask,request,render_template,jsonify
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
from flask_cors import CORS,cross_origin
```

```python
searchstring=input("Enter the product you want scrap data from:").replace(" ","")
```

```python
#Making the code general as later we can search for any object in flipkart to scrap data
flipkart_url="https://www.flipkart.com/search?q=" + searchstring
```

```python
#Here uReq is used to open the url provided. uReq is a alias for urlopen.
uclient=uReq(flipkart_url)
```

```python
#Here we are reading the html code of that opened page using read()
flipkart_page=uclient.read()
```

```python
#With the help of beautifulSoup we are beautifying the html data for reading properly
flipkart_html=soup(flipkart_page,"html.parser")
```

```python
flipkart_html
```

```python
#This will sort all the products in the current page using the class of html.
all_product=flipkart_html.findAll("div",{"class":"_1AtVbE col-12-12"})
```

```python
#There are almost 31 products in that current page.
len(all_product)
```

```python
#Selecting html code for an individual product
product1=all_product[0]
```

```python
product_href=product1.findAll("a")[0]["href"]
```

```python
productlink="https://www.flipkart.com" + product_href
```

```python
productlink
```

```python
uclient_product=uReq(productlink)
```

```python
product_page=uclient_product.read()
```

```python
product_html=soup(product_page,"html.parser")
```

```python
commentbox=product_html.find_all("div",{"class":"_16PBlm"})
```

```python
#Scraping the rating data from each comment
commentbox[4].findAll("div",{"class":"_3LWZlK _1BLPMq"})[0].text
```

```python
try:
    for i in commentbox:
        print(i.findAll("div",{"class":"_3LWZlK _1BLPMq"})[0].text)
except:
    pass        
        
```

```python
#scrapping  each comments from the page
commentbox[0].find_all("div",{"class":""})
```

```python
try:
    for i in commentbox:
        print(i.find_all("div",{"class":""})[0].text)
        print("\n")
except:
    pass        
```
